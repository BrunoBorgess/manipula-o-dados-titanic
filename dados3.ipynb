```python
import pandas as pd

base = pd.read_csv('titanic_train.csv')

base.head()
```

```python
# Verificar as infomações da base

base.info()
```

```python
# Eliminar a coluna Cabin 

base = base.drop('Cabin', axis=1) # eixo 1 signifca que quer eliminar uma coluna e eixo igual a 0 significa que quer eliminar uma linha
```

```python
base.info()
```

```python
# Elliminar valores vazios

base = base.dropna()
```

```python
base.info()
```

```python
# Selecionar as colunas do tipo 'object'

coluna = base.dtypes[base.dtypes.values == 'object'].index
print(coluna)
```

```python
# Eliminando essas colunas

base = base.drop(coluna, axis=1)
```

```python
base.info()
```

```python
# Definindo o x e y para treino
X = base.drop('Survived', axis=1)
y = base.Survived
```

```python
# importando KNN
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
# Documentação KNN https://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html
```

```python
# Criando nosso classificador
neigh = KNeighborsClassifier(n_neighbors=3) # Parametro Neigh classifica os valores aproximados por igual (exemplo: 1, 1, 1 e 0, irá classificar o 0 como 1)
```

```python
# Fazendo o fit com os dados / fit é entregar os dados ao modelo para que ele aprenda
neigh.fit(X, y)
```

```python
# Avaliando o modelo 
neigh.score(X, y)
```

```python
# Avaliando modelos de classficação com dados de teste
# titanic_teste.csv
# Devera ser feito todos os tratamentos novamente para realizar os testes com os dados de teste
```

```python
teste = pd.read_csv('titanic_test.csv')
teste.head()
```

```python
# Fazendo os mesmos tratamentos

teste = teste.drop('Cabin', axis=1)
teste = teste.dropna() # OPara eliminar as colunas desejadas
teste = teste.drop(coluna,axis=1)
```

```python
# Informações atualizadas

teste.info()
```

```python
# Separar o x e y da base
X_teste = teste.drop('Survived', axis=1)
y_teste = teste.Survived
```

```python
# Fazendo a predição com KNN

pred_KNN = neigh.predict(X_teste)
```

```python
# Importando a matriz de confusão 
from sklearn.metrics import confusion_matrix
```

```python
# Applicando a matriz para os valores em KNN
confusion_matrix(y_teste, pred_KNN)
```

```python
'''   0  1
    0 68 136  na primeira coluna 68 representa os numeros de zero que ele acertou, que são 68 verdadeiros e 136 errados, pois foram positovos, na segunda coluna ele previu 92 positivos e acertou e 35 negativos ele errou

    1 35 92'''
```

```python
# Verificando o erro  para KNN 
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_teste, pred_KNN)
```

```python
# Precisão
from sklearn.metrics import precision_score
```

```python
precision_score(y_teste, pred_KNN) # porcentagem de acerto de 40%
```

```python
# Recoll
from sklearn.metrics import recall_score # Recoll é a melhor opção para identificar fraudes
```

```python
recall_score(y_teste, pred_KNN)
```
